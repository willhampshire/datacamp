The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np

# Start coding here!
```

```python
nobel = pd.read_csv('data/nobel.csv')
print(nobel.head())
nobel.info()
```

```python
top_gender = str(nobel['sex'].value_counts().index[0])
top_country = str(nobel['birth_country'].value_counts().index[0])
print(f"{top_gender}\n{top_country}")
```

```python
nobel['decade'] = (nobel['year'] // 10)*10 # make decades col

countries_decades_count = nobel.groupby("decade")['birth_country'].count().astype(int)

usa_count = nobel[nobel['birth_country']=='United States of America'].groupby('decade')['birth_country'].count().astype(int)

ratio = usa_count / countries_decades_count

max_decade_usa = int(ratio.idxmax())
print(max_decade_usa)
```

```python
female_laurettes = nobel[nobel['sex']=='Female']\
.groupby(['category', 'decade'])['sex'].value_counts().astype(float)

male_laurettes = nobel[nobel['sex']=='Male']\
.groupby(['category', 'decade'])['sex'].value_counts().astype(float)

female_df = pd.DataFrame(female_laurettes).droplevel('sex')
male_df = pd.DataFrame(male_laurettes).droplevel('sex')

female_df.columns = ['count']
male_df.columns = ['count']
# print(female_df.head(), male_df.head())

both_sexes = female_df.merge(male_df, how='outer', suffixes=["_f", "_m"], on=['category', 'decade'])
both_sexes['ratio'] = both_sexes['count_f'] / both_sexes['count_m']

#print(both_sexes)

most_f_l = both_sexes.idxmax() # returns the max indexes in multi index for all 3 fields
# print(most_f_l)

cat, dec = most_f_l['ratio']
print(f"{cat}, {dec}")
max_female_dict = {dec: cat}

print(max_female_dict)
```

```python
first_woman = nobel[nobel['sex']=='Female']\
.sort_values(by='year')[['category', 'full_name']].iloc[0]

first_woman_name = first_woman['full_name']
first_woman_category = first_woman['category']

print(first_woman_name, first_woman_category)
```
